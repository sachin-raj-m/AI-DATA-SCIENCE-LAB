```python
import pandas as pd
import csv

#create a nested list
stud_list = [["ROLLNO","NAME","BRANCH","YEAR","CGPA"],
            [40,'Reshma','AD',2021,7.8],
            [39,'Rasal','AD',2021,8],
            [38,'Sharlet','AD',2021,10],
            [41,'Sachin','AD',2021,10],
            [37,'Naveen','AD',2021,8.5],
            [42,'Sakthi','AD',2021,9.5],
            [22,'Gouri','AD',2021,8.7],
            [21,'Gopika','AD',2021,9.2],
             ]

#opens a csv file
#create a csv writer object
#write data into csv file using writer
with open ('stud_details.csv', 'w', newline = '')as stud_file:
    writer = csv.writer(stud_file)
    writer.writerows(stud_list)
```
