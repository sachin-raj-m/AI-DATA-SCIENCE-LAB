```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (input feature and target)
x = np.array([15,23,25,28,33,37,42,45,51,58,60,63])
y = np.array([10,20,22,33,34,45,49,51,56,60,68,77])

plt.scatter(x,y,color = 'red',label='Datapoints')
```

```python
#Calculate the slope and intercept for the linear regression model
slope,intercept = np.polyfit(x,y,1)
```

```python
#Print Slope and intercept
print(f"Slope:{slope}")
print(f"Intercept:{intercept}")
```

```python
#Predict using the linear model
y_pred = slope*x+intercept
```

```python
#Plot the original data points and the linear regression line 
plt.scatter(x,y,color = 'red',label='Datapoints')
plt.plot(x,y_pred,color = 'Blue',label='Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
