```python
import pandas as pd
import numpy as np
data=pd.read_csv('heart_cleveland_upload.csv')
data.head(10)
```

```python
#calculate the correlation matrix
correlation_matrix=data.corr()
correlation_matrix
```

```python
#calculate the covarience matrix
covarience_matrix=data.cov()
covarience_matrix
```

```python
#identify the top 5 attributes closely related to the predicted attribute
target_correlation=correlation_matrix['condition'].abs().sort_values(ascending=False)
top_5_attributes=target_correlation[1:6]
print("The top 5 attributes closely related to the predicted attribute:")
print(top_5_attributes)
```
